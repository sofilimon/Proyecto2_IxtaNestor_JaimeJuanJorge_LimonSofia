## Descripción
En el siguiente proyecto se va a hacer un archivo con recomendaciones de distintas empresas en la bolsa. 
Usando de referencia la información de la página web YAHOO.
Este archivo va a tener la información de la bolsa desde el día 8 de noviembre, hasta la misma fecha del siguiente año,
de 5 diferentes empresas. 

En nuestro código vamos a definir una función que va a ayudarnos a predecir los 365 días que nosotros indiquemos 
cuando nosotros le ponemos el nombre de la empresa. 
Las empresas son DISNEY, NIKE, ADIDAS, NESTLE, BEYOND MEAT.

Objetivo:
1. Graficar cada una de las 5 empresas 
2. Poner las 5 en una sola grafica
3. Conclusión (elegir la mejor empresa para invertir en ciertas fechas)

Importar las librerías

```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import matplotlib.pyplot as plt
```

```python
def getCloses(names,start,end):
    datos=web.DataReader(names,"yahoo",start,end)
    return datos["Adj Close"]
```

```python
def simP(dato,start,ndays,N):
    R=dato.pct_change()
    mu=R.mean()
    std=R.std()
    Rsimulada=np.random.normal(mu,std,(ndays,N))
    fecha=pd.date_range(start,periods=ndays)
    RsimDF=pd.DataFrame(index=fecha,data=Rsimulada)
    Pproyectados=dato[-1]*(RsimDF+1).cumprod()
    return Pproyectados
```

```python
names=["DIS","NKE","ADDYY","BYND","NSRGY"]
start="2019-11-08"
end="2020-11-08"
datos=getCloses(names,start,end)
datos.head()
```

```python
Psim=simP(datos["DIS"],"2020-11-08",365,100)
Psim.plot(legend=False)
```

```python
UmbralD=datos["DIS"][-1]*1.2
B=Psim>UmbralD
#Prob=np.mean(B,axis=1)
Prob=B.mean(axis=1)
Prob.plot(grid=True)
```
